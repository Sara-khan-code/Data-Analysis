```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/menu.csv')
print(df)
```

```python
df.describe()
```

```python
columns = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
            'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
            'Iron (% Daily Value)']
df[columns].max()
```

```python
data = df[columns]
plt.figure(figsize = (10, 6))
sns.heatmap(data.corr())
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.set(style='whitegrid')
sns.boxplot(x= df['Calories'], y= df['Category'], data= df,hue= 'Category', width= 0.1)
plt.legend()
plt.show()
```

```python
for column in columns:
    maximum = df[column].max()
    item_name = df[df[column]==maximum]['Item'].iloc[0]
    print(f'{column}: {item_name}: {maximum}')
```

```python
sns .set(style='whitegrid')
plt.figure(figsize=(30,25))

plt.subplot(5, 2, 1)
sns.stripplot(data=df,x='Category',y='Calories',palette='bright')
plt.xticks(rotation=25,ha='right')
plt.title("Calories Content")

plt.subplot(5, 2, 2)
sns .stripplot(data=df,x='Category',y='Total Fat',palette='bright')
plt.title("Total Fat Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 3)
sns .stripplot(data=df,x='Category',y='Carbohydrates',palette='bright')
plt.title("Carbohydrates Content")
plt.xticks(rotation=20,ha='right')

plt.subplot(5, 2, 4)
sns .stripplot(data=df,x='Category',y='Dietary Fiber',palette='bright')
plt.title("Dietary Fiber Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 5)
sns .stripplot(data=df,x='Category',y='Sugars',palette='bright')
plt.title("Calories Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 6)
sns.stripplot(data=df,x='Category',y='Protein',palette='bright')
plt.title("Protein Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 7)
sns.stripplot(data=df,x='Category',y='Vitamin A (% Daily Value)',palette='bright')
plt.title("Vitamin A (% Daily Value) Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 8)
sns.stripplot(data=df,x='Category',y='Vitamin C (% Daily Value)', palette='bright')
plt.title("Vitamin C (% Daily Value) Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 9)
sns.stripplot(data=df,x='Category',y='Calcium (% Daily Value)', palette='bright')
plt.title("Calcium (% Daily Value) Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 10)
sns.stripplot(data=df,x='Category',y='Iron (% Daily Value)', palette='bright')
plt.title("Iron (% Daily Value) Content")
plt.xticks(rotation=25,ha='right')

plt.show()
```
